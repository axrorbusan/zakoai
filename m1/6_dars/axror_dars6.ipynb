```python
df = pd.read_csv('earthquakes.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop('alert', axis=1, inplace=True)
df.info()
```

```python
df.head()
#nice. i removed the column alert because it had too many empty values
```

```python
df.isnull().sum()
```

```python
df.drop(['cdi', 'dmin', 'felt', 'gap', 'mmi', 'nst'], axis=1, inplace=True)
df.isnull().sum()
```

```python
#nice. now i have some other columns with 1 value missing. let's fill them with something
print(df['mag'].dtype)
print(df['magType'].dtype)
print(df['tz'].dtype)
```

```python
for col in ['mag', 'magType', 'tz']:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
#nice. endi menda missing value lar yoq.
```

```python
import numpy as np
import datetime as dt
import pandas as pd
import requests

yesterday = dt.date.today() - dt.timedelta(days=1)
api = 'https://earthquake.usgs.gov/fdsnws/event/1/query'
payload = {
    'format': 'geojson',
    'starttime': yesterday - dt.timedelta(days=50),
    'endtime': yesterday
}
response = requests.get(api, params=payload)

response.status_code
```

```python
data_json = response.json()
data_json
```

```python
data_json.keys()
```

```python
data_json['features']
#this is where all the data is
```

```python
new_data = [data['properties'] for data in data_json['features']]
new_data
```

```python
df = pd.DataFrame(new_data)
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['time'] = pd.to_datetime(df['time'], unit='ms')
df['updated'] = pd.to_datetime(df['updated'], unit='ms')
df.head()
```

```python
df.to_csv('new_data.csv', index=False)
```
