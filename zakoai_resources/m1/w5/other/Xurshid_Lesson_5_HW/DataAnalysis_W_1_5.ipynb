### N-o'lchamli massivlar yaratishning boshqa usullari
`zeros` va `ones` funksiyalari yordamida massivlar yaratish

```python
arr4 = np.zeros((4, 4))
arr4
```

```python
arr5 = np.ones((4, 4)) # 2x4 o'lchamli barcha elementlari 1 dan iborat array
arr5
```

```python
# arange
arr6 = np.arange(1, 20, 3)
arr6
```

```python
# random
arr7 = np.random.rand(2, 3)
arr7
```

```python
# Normal distribution sampling
arr8 = np.random.randn(2, 3)
arr8
```

```python
arr9 = np.random.randint(8, size=(3, 3))
arr9
```
